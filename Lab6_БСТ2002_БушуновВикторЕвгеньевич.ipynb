```python
import networkx as nx
import matplotlib.pyplot as plt
import math

def convertMatrix(matrix):
    result_list = [(i, j, int(matrix[i][j]))
                   for j in range(len(matrix))
                   for i in range(len(matrix))
                   if matrix[i][j] != 'math.inf']

    return result_list

def floydWar(matrix, start, end):

    #меняем "math.inf" на math.inf
    V = [[int(matrix[i][j]) if matrix[i][j] != 'math.inf' else math.inf
          for j in range(len(matrix))] for i in range(len(matrix))]

    #число вершин в графе
    N = len(V)
    #двумерн список с помощью которго определяется кратчайший маршрут из одной вершины графа в другую
    P = [[v for v in range(N)] for u in range(N)]

    #сам алгоритм
    for k in range (N):
        for i in range(N):
            for j in range(N):
                d = V[i][k] + V[k][j]
                if V[i][j] > d:
                    V[i][j] = d
                    P[i][j] = k

    #кратчайший маршрут из одной вершины графа в другую
    path = [end]
    while end != start:
        end = P[end][start]
        path = [end] + path

    return path


with open('матрица.txt', 'r') as matrix:
    global graph
    graph = []

    for line in matrix:
        graph.append(line[:-1].split(', '))


# DiGraph — ориентированный граф
G = nx.DiGraph()
#заполняем граф (1,0,2)
G.add_weighted_edges_from(convertMatrix(graph))

#рисуем граф, подписи включены(лейблы) и вершины расположены по кругу
nx.draw(G, with_labels='True', pos=nx.circular_layout(G))
#делаем подписи (вес пути)
nx.draw_networkx_edge_labels(G, pos=nx.circular_layout(G), edge_labels=nx.get_edge_attributes(G, 'weight'))

path = floydWar(graph, 0, 7)
print(path)

#формируем список кортежей для вывода (1,0,2)
path = [(path[i], path[i+1]) for i in range(len(path)-1)]

#перерисовывем показывая путь
nx.draw_networkx_edges(G, pos=nx.circular_layout(G), edgelist=path, edge_color='red')
nx.draw_networkx_edge_labels(G, pos=nx.circular_layout(G), edge_labels=nx.get_edge_attributes(G, 'weight'))

plt.show()
```
